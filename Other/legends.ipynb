```python
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

SGR_TO_COLOR_AND_TEXT = {
    "basic": ["Basic", "#045275"],
    "moment20": ["Moment", "#089099"], 
    "filter10": ["Strata", "#7CCBA2"],
    "copula20": ["Copula-matching", "#FCDE9C"],
    "copula-filter10": ["Copula-strata", "#F0746E"], 
}

legend_elements = [Patch(facecolor=color, label=sgr) for (sgr, color) in SGR_TO_COLOR_AND_TEXT.values()]

fig, ax = plt.subplots(figsize=(20,1))
ax.legend(handles=legend_elements, loc='center', ncols=5, frameon=False, fontsize=20)
plt.axis("off")
plt.savefig("Legends/standard", bbox_inches="tight")
plt.show()
```

```python
COPULA_CONFIGS_TO_COLOR_AND_TEXT = {
    "copula-filter10": ["Load", "#F0746E"],
    "copula-filter-wind10": ["Wind onshore", "#4169E1"],
    "copula-filter-solar10": ["Solar PV", "#FFD700"],
    "copula-filter-combo10": ["Combo (Load + Wind onshore + Solar PV)", "#708090"]
}

legend_elements = [Patch(facecolor=color, label=sgr) for (sgr, color) in COPULA_CONFIGS_TO_COLOR_AND_TEXT.values()]

fig, ax = plt.subplots(figsize=(20,1))
ax.legend(handles=legend_elements, loc='center', ncols=4, frameon=False, fontsize=20)
plt.axis("off")
plt.savefig("Legends/copula-configs", bbox_inches="tight")
plt.show()
```

```python
from matplotlib.lines import Line2D

trials = [f"Trial {i}" for i in range(1, 6)]
legend_elements = [Line2D([0], [0], color=f"C{idx}", lw=4, label=trial) for (idx, trial) in enumerate(trials)]

fig, ax = plt.subplots(figsize=(20,1))
ax.legend(handles=legend_elements, loc='center', ncols=5, frameon=False, fontsize=20)
plt.axis("off")
plt.savefig("Legends/trials", bbox_inches="tight")
plt.show()
```

```python
import matplotlib.lines as mlines
import matplotlib.patches as mpatches

legend_elements = [mpatches.Patch(color=SGR_TO_COLOR_AND_TEXT["filter10"][1], label=f"{SGR_TO_COLOR_AND_TEXT["filter10"][0]} excl. the outlier"), mlines.Line2D([], [], color='red', marker='x', linestyle="None", label='The outlier', markersize=15)]

fig, ax = plt.subplots(figsize=(20,1))
ax.legend(handles=legend_elements, loc='center', ncols=2, frameon=False, fontsize=20)
plt.axis("off")
plt.savefig("Legends/strata-outlier", bbox_inches="tight")
plt.show()
```

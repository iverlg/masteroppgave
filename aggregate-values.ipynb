### Aggregate countries to regions

Sum load across regions

```python
import pandas as pd 

df_load = pd.read_csv("copula_testing/scenario_data/electricload.csv")

# Based on Stian paper
north = ["DK", "FI", "SE", "NO1", "NO2", "NO3", "NO4", "NO5", "GB", "IE", "EE", "LV", "LT"]
west = ["BE", "ES", "FR", "CH", "NL", "PT", "IT"]
east = ["AT", "BG", "LU", "CZ", "DE", "GR", "HR", "HU", "MK", "PL", "RO", "RS", "SI", "SK", "BA"]

df_load
```

```python
df_load["North"] = df_load[north].sum(axis=1)
df_load["West"] = df_load[west].sum(axis=1)
df_load["East"] = df_load[east].sum(axis=1)
df_load = df_load.drop(columns=north+west+east)
df_load
```

```python
df_load.to_csv("copula_testing/scenario_data-aggr/electricload.csv")
```

Representative countries for renewables sampling

```python
north_country = "GB"
west_country = "FR"
east_country = "DE"

renewable_sources = ["windonshore", "solar", "windoffshore", "hydroror", "hydroseasonal"]

for rs in renewable_sources:
    _df = pd.read_csv(f"copula_testing/scenario_data/{rs}.csv")
    _df["North"] = _df[[north_country]]
    _df["West"] = _df[[west_country]]
    _df["East"] = _df[[east_country]]

    drop_cols = set(north+east+west).intersection(_df.columns)
    _df = _df.drop(columns=drop_cols)
    _df.to_csv(f"copula_testing/scenario_data-aggr/{rs}.csv")
```
